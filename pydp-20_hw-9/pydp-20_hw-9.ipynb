## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'АВ222С96'`

Результат: 
`Номер АВ222С валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
def valid_trans_no(trans_no):
    '''
    Функция принимает на вход строку и проверяет, является ли она валидным транспортным номером.
    Если номер валиден, функция возвращает отдельно номер и регион.
    '''
    
    regular_trans_no = r'^([АВЕКМНОРСТУХ]{1})(\d{3})([АВЕКМНОРСТУХ]{2})(\d{2,3})$'
    result = re.search(regular_trans_no, trans_no)
    if result is not None:
        return f'Номер {result[1]+result[2]+result[3]} валиден. Регион: {result[4]}'
    else:
        return 'Номер не валиден'
````

````python
car_id = 'А567ЕУ190'

valid_trans_no(car_id)
````

## Задание 2

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:  

`some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'`

Результат: 
`Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.`

````python
def remove_repeats(input_string):
    '''
    Используя рекурсию, функция удаляет все последовательные повторы слов из заданной строки с помощью регулярных выражений.
    '''
    if re.findall(r'(\s\w+)\1', input_string)==[]:
        return input_string
    else:
        edited_string = re.sub(r'(\s\w+)\1', r'\1', input_string)
        return remove_repeats(edited_string)
````

````python
some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'

remove_repeats(some_string)
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
def acronym_maker(some_string):
    '''
    Функция возвращает акроним по переданной в нее строке со словами.
    '''
    return re.sub(r'(\w)(\w+\s{0,1})', r'\1', some_string).upper()

some_words = 'Информационные технологии'
print(acronym_maker(some_words))

some_words = 'Near Field Communication'
print(acronym_maker(some_words))
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', xyz@gmail.com]`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
def domain_distribution(emails):
    email_domains = []

    for email in emails:
        email_domains.append(re.search(r'(?:.+@)(\w+\.\w+)', email)[1])

    domains = set(email_domains)
    for email_domain in domains:
        print(f'{email_domain}: {email_domains.count(email_domain)}')

emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
domain_distribution(emails)
````
